Formulate a hypothesis related to the dataset.
Conduct a statistical test (e.g., t-test) to test the hypothesis.
Provide conclusions based on the test results.


We suppose that there is no difference in rating between cold and hot cereal types (H0).

```python
cold = df[df['type']=='C']
hot = df[df['type']=='H']
stats.ttest_ind(cold['rating'], hot['rating'])
```

Our p value is larger than 0.05 that means we can't defute hypothesis about no difference between the types of cereal.

```python
df['type'].value_counts()
```

As we can see the hot cereals are underrepresented in our dataset which could (most probably has) lead to the result of our t-test

#### Identify and handle outliers in the dataset.
Explain the methodology and reasoning behind your approach.


```python
#check for missing values
df.isna().sum()
#none are found
```

```python
df.describe()
```

As you can see there are negative values in carbo, sugars and potassium. Our suggestion would be to set them to zero.

```python
negative_values_mask = (df['carbo'] < 0) | (df['sugars'] < 0) | (df['potass'] < 0)

# Set negative values in 'carbo', 'sugars', and 'potassium' columns to zero
df.loc[negative_values_mask, ['carbo', 'sugars', 'potass']] = 0
```

### Correlation Analysis:
Compute and interpret the correlation between two relevant variables in the dataset.

```python
correlation_matrix = df.corr(numeric_only=True)

# heatmap of the correlations
plt.figure(figsize = (8, 6))
sns.heatmap(correlation_matrix, annot = True, cmap = "coolwarm", fmt= '.2f')
plt.title("Correlation Matrix")
plt.show()
```

We can see that the correlation between sugars in a serving and rating of a cereal are negatively correlated (I was expecting different result). That means that cereal with a sweeter taste are usually also less liked.
